```python
#import libraries
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#create the dataframe
raw_csv_data=pd.read_csv("IMDB-Movie-Data.csv")
raw_csv_data
```

```python
df=raw_csv_data.copy()
```

```python
# To see all columns&rows
pd.options.display.max_columns=None
pd.options.display.max_rows=None
display(df)
```

```python
#To get information about dataset
df.info()
```

```python
#checking missing values in the dataset
df.isnull()
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull())
plt.show()
```

```python
len(df)
```

```python
#To get missing value percentage
percentage_missing=df.isnull().sum()*100/len(df)
percentage_missing
```

```python
#dropping all missing values
df.dropna(axis=0).shape
```

```python
#check for duplicate data
duplicate_data=df.duplicated().any()
duplicate_data
```

```python
# To get statistics about the dataframe
df.describe(include='all')
```

```python
df.columns
```

```python
#display title of movie having runime >=180minutes
df[df['Runtime (Minutes)']>=180]['Title']
```

```python
#in which year there was the highest average voting
df.groupby('Year')['Votes'].mean().sort_values(ascending=False)
```

```python
sns.barplot(x='Year',y='Votes',data=df)
plt.title("Average Votes by Year")
plt.show()
```

```python
#in which year there was the highest average revenue
df.groupby('Year')['Revenue (Millions)'].mean().sort_values(ascending=False)
```

```python
sns.barplot(x='Year',y='Revenue (Millions)',data=df)
plt.title("Highest Average Revenue by Year")
plt.show()
```

```python
#To get the average rating for each director
df.groupby('Director')['Rating'].mean().sort_values(ascending=False)
```

```python
#To display top10 lengthy movies Tiltle and Runtime
top10 = df.nlargest(10,'Runtime (Minutes)')[['Title','Runtime (Minutes)']].set_index('Title')
top10
```

```python
sns.barplot(x='Runtime (Minutes)',y=top10.index,data=top10)
plt.show()
```

```python
#To get no.of movies per Year
df['Year'].value_counts()
```

```python
sns.countplot(x='Year',data=df)
plt.show()
```
